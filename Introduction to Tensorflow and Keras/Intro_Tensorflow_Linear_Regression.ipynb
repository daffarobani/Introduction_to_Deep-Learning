```python
#Importing modules
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(101)
tf.set_random_seed(101)
```

```python
#Create 50 data points 
x = np.linspace(0,50,50)
#Add noise to y
y = 0.3*x + 2 + np.random.uniform(-2,2,50)
```

```python
#Visualize the training data
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Training Data')
plt.show()
```

```python
#Define variable X and Y in tensorflow
X = tf.placeholder('float')
Y = tf.placeholder('float')
```

```python
#Define parameter W and b
W = tf.Variable(np.random.randn(), name='W')
b = tf.Variable(np.random.randn(), name='b')
```

```python
#Define the hyperparameter of the model, learning rate and the number of epochs
learning_rate = 0.01
training_epochs = 1000
```

```python
n = len(x)
#Hypothesis
y_pred = tf.add(tf.multiply(X, W), b)
#Mean Squared Error Cost Function
cost = tf.reduce_sum(tf.pow((y_pred-Y),2))/(2*n)
#Gradient Descent Optimizer
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

#Global Variables Initializer
init = tf.global_variables_initializer()
```

```python
#Training process inside Tensorflow Session
with tf.Session() as sess:
    #Initializing Vatiables
    sess.run(init)
    
    #Iterating through all the epochs
    for epoch in range(training_epochs):
        #Feeding each data point into the optimizer
        for(_x,_y) in zip(x,y):
            sess.run(optimizer,feed_dict={X:_x,Y:_y})
            
            #Displaying the result after every 50 epochs
            if (epoch+1) %50 ==0:
                #Calculating the cost at every epoch
                c = sess.run(cost, feed_dict={X:x,Y:y})
                print("Epoch {}, cost= {}, W={}, b={}".format((epoch+1),c,sess.run(W),sess.run(b)))
                #print("Epoch", (epoch+1), ": cost=", c, "W=". sess.run(W),'b=', sess.run(b))
    #Storing necessary values to be used outside the function
    training_cost = sess.run(cost, feed_dict={X:x,Y:y})
    weight = sess.run(W)
    bias = sess.run(b)
```

```python
print("Training cost= {}, Weight = {}, bias = {}".format(training_cost,weight,bias))
#Calculating the predictions
predictions = weight*x+bias

#Plotting predictions
plt.scatter(x,y,label='Original Data')
plt.plot(x,predictions, label='Fitted Line')
plt.title('Linear Regression Result')
plt.legend()
plt.show()
```
